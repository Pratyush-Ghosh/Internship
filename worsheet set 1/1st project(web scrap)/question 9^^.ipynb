```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url='https://www.dineout.co.in/delhi-restaurants/buffet-special'
```

```python
soup = requests.get(url)
buffet=BeautifulSoup(soup.content,'html.parser')
```

```python
food=buffet.find_all('div',{'restnt-card restaurant'})
```

```python
name=[]
cuisine=[]            
location=[]
rating=[]
image=[]
```

```python
for i in food:
    name.append(i.a.text)
    cuisine.append(i.find('div',{'class','detail-info'}).text.split('|')[1])
    location.append(i.find('div',{'class':'restnt-loc ellipsis'}).text)
    rating.append(i.find('div',{'class':'restnt-rating rating-4'}).text)
```

```python
for j in buffet.find_all('img',{'class':'no-img'}):
    image.append(j['data-src'])
```

```python
data=list(zip(name,cuisine,location,rating,image))
```

```python
df=pd.DataFrame(data,columns=['Restaurant Name','Cuisine','location','rating','image url'])
```
