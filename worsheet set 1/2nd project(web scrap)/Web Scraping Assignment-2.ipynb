Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You

Q5: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link:https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYL0BETT&marketplace=FLIPKART

```python
url='https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYL0BETT&marketplace=FLIPKART&page=13'
driver.get(url)#url
```

```python
rating=[]
summary=[]
full_summary=[]
for i in range(11): #range for number of pages it's gonna scrape
    tt=driver.find_elements_by_xpath('//div[@class="_3LWZlK _1BLPMq"]') #storeing the path in obj
    yy=driver.find_elements_by_xpath('//p[@class="_2-N8zT"]')
    mm=driver.find_elements_by_xpath('//div[@class="t-ZTKy"]')
    for d in tt: #loop for company name
        rating.append(d.text)
    for f in yy: #loop for summary
        summary.append(f.text)
    for o in mm: #loop for review 
        full_summary.append(o.text)
    driver.find_element_by_xpath('//a[@class="_1LKTO3"]').click() #path of next page for loop
    sleep(2) #wait for 2sec to rescraping the data
```

```python
data=list(zip(rating,summary,full_summary))[:100] #zipping top 100 review
df=pd.DataFrame(data,columns=("Rating","Summary","Full Review"))
df
```

```python
driver.find_element_by_xpath('//*[@id="salaries"]/main/section[1]/div[2]/div[1]/span/div/div/div[1]').click() 
#clicking on first suggestion
```

```python
company=[]
salaries=[]
for i in driver.find_elements_by_xpath('//div[@class="name"]'): #loop for salaries and comapny name
    company.append(i.text.split()[1])
    salaries.append(i.text.split()[-2])
exp=[]
for k in driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]'): #loop for experience
    exp.append(k.text.split()[-3])
minsal=[]
avgsal=[]
maxsal=[]
for j in driver.find_elements_by_class_name("result-row"): #loop for minimun, average, maximum salary in page
    minsal.append(j.text.split()[-3])
    avgsal.append(j.text.split()[-5])
    maxsal.append(j.text.split()[-1])
```
